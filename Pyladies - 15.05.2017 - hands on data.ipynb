supervised vs unsupervised machine learning
supervised (nadzorowane uczenie) - wyjścia są określone, np. maile typu 'spam' i 'niespam'
unsupervised (nienadzorowane uczenie) - nie wiemy czego spodziewać się na końcu, nie ma określonego wyjścia, np. wykrywanie nieokreślonych wzorów (patterns) w danych; np. klastrowanie
